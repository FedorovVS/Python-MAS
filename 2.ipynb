### Массивы  
Более функциональный вариант списка предоставляет библиотека NumPy. Это удобный инструмент при работе с любого рода вычислениями. Алгоритмы, зашитые внутрь, реализованы достаточно эффективно, поэтому NumPy применяется практически везде.  
Любая библиотека ставится так:

````python
import numpy as np
````

np - краткое общепринятое обозначение для этой библиотеки, все популярные инструменты имеют такую нотацию, не стоит от неё отходить.  
Массивы в NumPy не ограничены в размерности и обычно хранят в себе числа (int, float, etc). Создание массива:

````python
a = np.zeros(5) #  массив из 5 нулей
b = np.ones([3, 3]) # матрица из единиц 3х3
c = np.array([1, 3, 4, 8]) # массив из списка
d = np.linspace(0, 10, 50) # начало, конец, количество точек. Удобно при рисовании функций. С дополнительными параметрами и похожими функиями можно ознакомиться в документации
e = np.arange(0, 10, 0.1) #  начало, конец, шаг. Аналог range()
````

NumPy позволяет проводить преобразования элементов массивов очень легко, так можно получить значения квадратичной функции в точках, перечисленных в массиве:

````python
f = d**2
````

Также в библиотеке содержатся все необходимые математические операции: логарифмы, среднее, стандартное отклонение, умножение матриц и многое другое.

### Настройка рисунка  
Читаемость ваших графиков - залог успеха. Нужно снабжать их подписями, делать красивую разметку и раскрашивать невырвиглазно. Далее приведён базовый функционал для этого. Иногда к нему придётся прибегать и при применении более высокоуровневых библиотек.

```
plt.xlabel() #  подпись оси х
plt.ylabel() # подпись оси у
plt.xticks() # куда ставить деления, как подписать (labels)
plt.yticks() 
plt.title() #  название графика
plt.grid(True) #  сетка координат
```
Цвет можно настроить передачей в функцию графика аргумента-строки типа 'CN', где N = 0..9. При одновременном отображении нескольких графиков цвет может изменяться автоматически.  
Далее приведён некоторый содержательный пример графика:

````python
x = np.arange(-10, 10.01, 0.01)
plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), label=r'$f_1(x)=\sin(x)$')
plt.plot(x, np.cos(x), label=r'$f_2(x)=\cos(x)$')
plt.plot(x, -x, label=r'$f_3(x)=-x$')
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.grid(True)
plt.legend(loc='best', fontsize=12)
plt.savefig('figure_with_legend.png')
plt.show()
````

### Немного ООП
В мире ООП есть ряд базовых вещей: всё что угодно, с чем осуществляется взаимодействие - объект. Каждый объект принадлежит классу, это описание для похожих объектов. У объекта есть набор параметров и методов (функций). Все станет ясно из примера:

````python
class Astronomer:
    def __init__ (self, name, years, region, telescope):
        self.name = name
        self.years = years
        self.region = region
        self.telescope = telescope
        self.xp = 100
        self.info = 0
    
    def solve_problem (self):
        self.xp -= 5
        self.info += 1
    
    def solve_teor (self):
        for i in range(5):
            solve_problem()
    
    def get_telescope (self):
        self.xp += 10
        self.telescope = True
        self.info += 5
    
    def print_info (self):
        print(self.name, self.years, self.xp, self.info)
````

````python
astro = Astronomer('Ivan', 24, 'msk', True)
````

````python
astro.print_info()
````

````python
astro.solve_problem()
````

````python
astro.print_info()
````
